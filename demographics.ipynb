# Demographics extension

This extension allows you to add a demographics page to your survey in a single line of code. It includes 20 demographics items based on the World Values Survey.

You can add one of several default "bundles" of items such as `basic_demographics` and `comprehensive_demographics` or customize your demographics page with `demographics`.

Check out <a href="https://dsbowen.github.io/hemlock-demographics/" target="_blank">the docs</a> for more details!

```python
from hemlock import push_app_context

app = push_app_context()
```

```python
from hemlock_demographics import basic_demographics

basic_demographics(page=True).preview(filename='basic_demographics')
```

```python
from hemlock_demographics import comprehensive_demographics

comprehensive_demographics(page=True).preview(filename='comprehensive_demographics')
```

```python
from hemlock_demographics import demographics

demographics('birth_year', 'education', 'employment', 'sector', page=True).preview(filename='demographics')
```
